```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
from scipy.stats import f as fisher_f
```

```python
df = pd.read_csv("./../data/Raw Data_GeneSpring.txt",delimiter='\t')
# removing the all zero rows      # df['total_sum'] = df.iloc[:,1:50].sum(axis=1)
# df = df.drop([ 3420-2,  3850-2,  5073-2,  8420-2, 13607-2, 36194-2, 36271-2],axis=0)

A_ = np.zeros(shape=(48,4))
A_[0:12] = [1,0,1,0]
A_[12:24] = [1,0,0,1]
A_[24:36] = [0,1,1,0]
A_[36:48] = [0,1,0,1]

A = np.zeros(shape=(48,4))
A[0:12] = [1,0,0,0]
A[12:24] = [0,1,0,0]
A[24:36] = [0,0,1,0]
A[36:48] = [0,0,0,1]
# df.size
```

```python
n1 = np.matmul(np.matmul(A,np.linalg.pinv(np.matmul(A.T,A))),A.T)
n2 = np.matmul(np.matmul(A_,np.linalg.pinv(np.matmul(A_.T,A_))),A_.T)
num = np.subtract(n1,n2)
den = np.subtract(np.identity(n1.shape[0]),n1)
rankA = np.linalg.matrix_rank(A)
rankA_= np.linalg.matrix_rank(A_)
# print(rankA,rankA_)
# print(n1.shape,n2.shape,num.shape,den.shape)
```

```python
def func(x):
    h = np.array(x[1:49])
    n = np.matmul(np.matmul(h.T,num),h)
    d = np.matmul(np.matmul(h.T,den),h)
    return n / d * (48 - rankA) / (rankA - rankA_)if d else 0
    
# df.drop([3418,3419],axis=0)
df ['F-stat'] = df.apply(func,axis=1)
df1 = rankA - rankA_
df2 = 48 - rankA
df['p-val'] = 1 - fisher_f(df1,df2,0).cdf(df['F-stat'])
```

```python
df['p-val'].hist()
plt.show()
```

```python
symbols = np.array(list(map(str,df[df['p-val']<0.05]['GeneSymbol'])))
```

```python
def boxplot_for_sanity_check(geneSymbol):
    fig, ax = plt.subplots()
    ax.set_title(geneSymbol)
    ax.set_xticklabels(['MNs','MS','FNs','FS'])
    aadac = np.array(df[df['GeneSymbol'] == geneSymbol])
    data = [list(map(float,aadac[0,1:13])),list(map(float,aadac[0,13:25])),list(map(float,aadac[0,25:37])),list(map(float,aadac[0,37:49]))]
    # print(data)
    ax.boxplot(data)
    plt.show()
```

```python
import copy
filenames = ["XenobioticMetabolism1",
            "FreeRadicalResponse",
            "DNARepair1",
            "NKCellCytotoxicity"]
x = []
for filename in filenames:
    symbol_ = copy.deepcopy(symbols)
    pathname = "./../data/"+filename+".txt"
    Xm = pd.read_csv(pathname,delimiter='\t')
    Xm_np = np.array(list(map(str,list(Xm.iloc[1:,0]))))
    print(Xm_np)
    Xin = np.intersect1d(symbol_,Xm_np)
    print("intersection with "+filename+" is "+str(Xin.size) )
    for genesymbol in Xin:
        boxplot_for_sanity_check(genesymbol)
    for gene in Xin:
        x_ = list(df[df['GeneSymbol'] == gene])
        
        
```
